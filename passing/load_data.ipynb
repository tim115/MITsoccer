## b. Line-ups data

```python
import pandas as pd
import json

with open('../data/statsbomb360/lineups/3788747.json', 'r') as f:
    lineups = json.load(f)

lineups_df = pd.json_normalize(lineups, record_path=[['lineup']])
lineups_df
```

## c. Three-Sixty data

```python
import pandas as pd
import json

path = '../data/statsbomb360/three-sixty/3788747.json'

with open(path, 'r') as f:
    data = json.load(f)

freeze_frame = pd.json_normalize(data, record_path=[['freeze_frame']])
visible_area = pd.json_normalize(data, record_path=[['visible_area']])
df_raw = pd.read_json(path)
df_raw, freeze_frame, visible_area
```
